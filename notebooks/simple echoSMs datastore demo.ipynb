```python
import requests
from IPython.display import Image, display
from rich.pretty import pprint
```

The echoSMs datastore web API is available on a publically-available prototype server or you can run your own web API on your hackathon server.

You'll need to manually start the hackathon server one using these steps:

- Open a terminal (use the Jupyterlab launcher to do this - accessible via the `+` symbol to the right of the tabs)
- Change to the /data_store_testing directory - type `cd /data_store_testing`
- Run the web API server - type `fastapi dev`

The code below should then work.

```python
# The URL of the web API. This can either be the API available on
# your hackathon server or the publically-available echoSMS prototype

# The one on your hackathon server
api_URL = 'http://127.0.0.1:8000'
# The prototype echoSMs one
api_URL = 'https://echosms-data-store-app-ogogm.ondigitalocean.app'

# Get a list of all specimens in the datastore.
d = requests.get(api_URL + '/v2/specimens')
specimens = d.json()

# Print out how many specimens were provided
print(len(specimens))

# Print out the metadata for the first specimen in the list. There is no shape data in this
pprint(specimens[0])
```

```python
# Get the shape data for one specimen
d = requests.get(api_URL + '/v2/specimen/CLAY_HORNE_B/data')

# Print out the specimen data. Metadata and shape data are provided. 
pprint(d.json())
```

```python
# Get a plot of the shape for one of the specimens
d = requests.get(api_URL + '/v2/specimen/CLAY_HORNE_B/image')

# A couple of hoops to jump through to show the plot in a jupyter notebook
display(Image(d.content))
```
